```python
import dill

import numpy as np
import pandas as pd

import tensorflow as tf

from tensorflow import keras
```

```python
cnn_model = keras.models.load_model('./beijing_air_2_5_cnn_model.h5')
dnn_model = keras.models.load_model('./beijing_air_2_5_dnn_model.h5')
rnn_model = keras.models.load_model('./beijing_air_2_5_rnn_model.h5')

with open('./beijing_air_2_5_test_data.dill', 'rb') as f:
    dataset_test = dill.load(f)
```

```python
import sys

sys.path.insert(0, './../../src/')

from tsmule.xai.lime import LimeTS
```

```python
def predict_(x):
    prediction = cnn_model.predict(x[np.newaxis])[0]
    return prediction

explainer = LimeTS()
sample = dataset_test[0][0]
explainer.explain(sample, predict_)
```

```python
def predict_(x):
    prediction = dnn_model.predict(x[np.newaxis])[0]
    return prediction

explainer = LimeTS()
sample = dataset_test[0][0]
explainer.explain(sample, predict_)
```

```python
def predict_(x):
    prediction = rnn_model.predict(x[np.newaxis])[0]
    print(prediction)
    return prediction

explainer = LimeTS()
sample = dataset_test[0][0]
explainer.explain(sample, predict_)
```
